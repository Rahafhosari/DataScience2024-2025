```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer,make_column_transformer,make_column_selector
from sklearn import set_config
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,BaggingClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
set_config(transform_output='pandas')
```

```python
path = '/content/drive/MyDrive/AXSOSACADEMY/02-IntroML/Week07/23122024 - Parkinson Exercise/pd_speech_features.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.describe()
```

```python
df['class'].value_counts()
```

Train Split Test

```python
y = df['class']
X = df.drop(columns='class')

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

## Decision Tree

```python
dec_tree = DecisionTreeClassifier(random_state=42)
dec_tree.fit(X_train,y_train)
```

Predictions

```python
train_pred_dec_tree = dec_tree.predict(X_train)
test_pred_dec_tree = dec_tree.predict(X_test)
```

```python
train_score_dec_tree = accuracy_score(y_train,train_pred_dec_tree)
test_score_dec_tree = accuracy_score(y_test,test_pred_dec_tree)

print(f'Train Score: {train_score_dec_tree}')
print(f'Test Score: {test_score_dec_tree}')
```

The model is overfitted.

## Bagging Tree

```python
bag_tree = BaggingClassifier(random_state=42)
```

```python
bag_tree.fit(X_train,y_train)
```

```python
train_pred_bag_tree = bag_tree.predict(X_train)
test_pred_bag_tree = bag_tree.predict(X_test)

train_score_bag_tree = accuracy_score(y_train,train_pred_bag_tree)
test_score_bag_tree = accuracy_score(y_test,test_pred_bag_tree)

print(f'Train Score: {train_score_bag_tree}')
print(f'Test Score: {test_score_bag_tree}')
```

Still overfitted.

## Random Forest

```python
rand_forest = RandomForestClassifier(random_state=42)
rand_forest.fit(X_train,y_train)
```

```python
train_pred_rand_forest = rand_forest.predict(X_train)
test_pred_rand_forest = rand_forest.predict(X_test)
```

```python
train_score_rand_forest = accuracy_score(y_train,train_pred_rand_forest)
test_score_rand_forest = accuracy_score(y_test,test_pred_rand_forest)

print(f'Train Score: {train_score_rand_forest}')
print(f'Test Score: {test_score_rand_forest}')
```

## Tuning Random Forest Model

```python
rand_forest.get_params()
```

```python
rand_forest_params = {'bootstrap': [True,False],
                      'max_depth': [0,5,10,20,None],
                      'n_estimators':[10,20,50,100]
                      }
```

```python
# Instantiate the gridsearch
gridsearch = GridSearchCV(rand_forest, rand_forest_params, n_jobs=-1, cv = 5, verbose=1)
# Fit the gridsearch on training data
gridsearch.fit(X_train, y_train)
```

We can't decide which model is better since there's an overfit in all of them.
